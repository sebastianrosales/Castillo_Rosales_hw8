```python
# primera  opcion de correr puntos
dU_x=zeros(len(latitud)-1)
dU_y=zeros(len(latitud)-1)
dX_x=zeros(len(latitud)-1)
dX_y=zeros(len(latitud)-1)
for  i in range (len(latitud)-1):
    dU_x[i]=(longitud[i]-longitud[i+1])-(este[i]-este[i+1])
    dU_y[i]=(latitud[i]-latitud[i+1])-(norte[i]-norte[i+1])
    #el dX para el tensor debe ser la resta de los puntos  originales al origen, es decir, la distacia  entre  los  puntos  que  comparamos 
    dX_x[i]=dis_origenx[i+1]-dis_origenx[i]
    dX_y[i]=dis_origeny[i+1]-dis_origeny[i]
           
    # grafica 
    scatter( dX_x[i],dU_x[i], c='r' )# pregunta
    scatter( dX_y[i],dU_y[i], c='b' )

    #scatter(dU_x[i], dis_origenx[i], c='r' )# pregunta
    #scatter(dU_y[i], dis_origeny[i], c='b' )
title("comparacion dU Vs dX Para  todas las  estaciones")
xlabel("dX")
ylabel("dU")
#plot(dU_x, dX_x)
```

```python
s=((len(latitud)-1),2)
dU=zeros(s)
dX=zeros(s)
a=(2,2)
valoresPropios=zeros(s)
for  i in range (len(latitud)-1):
    dU[i,0]=(longitud[i]-longitud[i+1])-(este[i]-este[i+1])
    dU[i,1]=(latitud[i]-latitud[i+1])-(norte[i]-norte[i+1])
    #el dX para el tensor debe ser la resta de los puntos  originales al origen, es decir, la distacia  entre  los  puntos  que  comparamos 
    dX[i,0]=dis_origenx[i+1]-dis_origenx[i]
    dX[i,1]=dis_origeny[i+1]-dis_origeny[i]
    # Matrix de transformacion (Elongacion) 
    ejk=zeros(a)
    ekj=zeros(a)
    E=zeros(a)
    for j in range(2):
        for k in range(2):
            ejk[j,k]=dU[i,j]/dX[i,k] # matris de elongacion - No simetrica 
            ekj[j,k]=dU[i,k]/dX[i,j]  
            #eigenvalues, eigenvectors = eig(ei)
            
            E[j,k]=(ejk[j,k]+ekj[j,k])/2.0 # Matriz simetrica 
    
    eigenvalues, eigenvectors = eig(E) # valors y vectores  propios  de la matriz 

    arrow(dX[i,0], dX[i,1], eigenvectors[0,0] , eigenvectors[0,1] , color=(0,0,1))
    arrow(dX[i,0], dX[i,1], eigenvectors[0,0] , eigenvectors[1,0] , color =(1,0,0))# son  iguales  con sentido contrario. 
    
    valoresPropios[i,0]=eigenvalues[0]
    valoresPropios[i,1]=eigenvalues[1]
       
xlim(-2,2)
ylim(-2,2)
figsize(10,10)        
title("comparacion autoverctores de matriz simetrica Para  todas las  estaciones")
xlabel("dX")
ylabel("V")
# imprime  solo el  ultimo 
#print E
#eigenvalues, eigenvectors = eig(E)
#for i in range(2):
    #print "eigenvalue %d"%i, eigenvalues[i]
    #print "eigenvector %d"%i, eigenvectors[:,i]
    #print "\n"
```
