## Procesamiento de las imagenes

El objetivo es encontrar el centro y los puntos que describen el
contorno de las ves\'iculas. Para esto vamos a tomar que desde el
centro de la ves\'icula el contorno cumple la siguiente ecuaci\'on en
coordenadas polares:

\begin{equation}
r(\theta) = r_{0} + r_{1}\sin(m\theta + \phi), 
\label{eq:1}
\end{equation}
donde $r_0$, $r_1$, $\phi$ y $m$ son par\'ametros libres, con $m>0$ entero.

Deben escribir un programa que a trav\'es de MCMC encuentre las
distribuciones de probabilidad de la posici\'on del centro del
c\'irculo y de los par\'ametros $r_0$, $r_1$, $\phi$ y $m$. 
Para esto la idea es que en cada paso se genere una imagen para ser
comparada con la imagen original y poder calcular el $\chi^2$.

```python
def modelo(theta,r0,r1,m, phi):
    return r0+r1*cos(m*theta+phi)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2/var(y_obs))
    #print "varianza= "+str(var(y_obs))
    #print "chi= " + str(chi_squared)    
    #print "l= " + str(exp(-chi_squared)) 
    return exp(-chi_squared)
```

#### ($\star$)Es importante notar que se hizo un corrimiento del centro de coordenadas para que se pueda ajustar al modelo propuesto.

## MCMC

```python
r0_walk = empty((0))
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
l_walk = empty((0))
l_inst=0.0

#Este While se asegura de no quedar atrapado en l=0
while l_inst==0.0:
    
    a=((random.random()*10)+95)
    b=((random.random())+1)*5
    c=random.random()
    d=(random.random()-0.5)*pi/6
    r_inst=modelo(theta_obs, a,b,c,d)
    l_inst=likelihood(r_obs,r_inst)


r0_walk = append(r0_walk,a)
r1_walk = append(r1_walk,b)
m_walk = append(m_walk, c)
phi_walk = append(phi_walk, d)


r_init = modelo(theta_obs, r0_walk[0], r1_walk[0],m_walk[0],phi_walk)
l_walk = append(l_walk, likelihood(r_obs, r_init))
print "r0= "+str(r0_walk[0])+", r1= "+str(r1_walk[0])+", m= " +str(m_walk[0])+ ", phi= "+str(phi_walk[0])
print l_walk 
scatter(theta_obs,r_obs,s=5)
scatter(theta_obs,r_init,c="r")
xlim (-pi,pi)
ylim(80, 120)
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    #print "i= "+str(i)
    r0_prime = np.random.normal(r0_walk[i], 0.1) 
    r1_prime = np.random.normal(r1_walk[i], 0.1)
    m_prime = np.random.normal(m_walk[i], 0.1)
    phi_prime = np.random.normal(phi_walk[i], 0.1)
    
    r_init = modelo(theta_obs, r0_walk[i], r1_walk[i],m_walk[i],phi_walk[i])
    r_prime = modelo(theta_obs, r0_prime, r1_prime,m_prime,phi_prime)
    
    l_prime = likelihood(r_obs, r_prime)
    
    l_init = likelihood(r_obs, r_init)
    alpha = l_prime/l_init
    
    if(alpha>=1.0):
        r0_walk = append(r0_walk, r0_prime)
        r1_walk = append(r1_walk, r1_prime)
        m_walk = append(m_walk, m_prime)
        phi_walk = append(phi_walk, phi_prime)
        l_walk = append(l_walk, l_prime)
        #print "Dentro del primer if"
    else:
        """beta = random.random()"""
        if(beta<=alpha):
            r0_walk = append(r0_walk, r0_prime)
            r1_walk = append(r1_walk, r1_prime)
            m_walk = append(m_walk, m_prime)
            phi_walk = append(phi_walk, phi_prime)
            l_walk = append(l_walk, l_prime)
            #print "Dentro del segundo if"
        else:

            r0_walk = append(r0_walk, r0_walk[i])
            r1_walk = append(r1_walk, r1_walk[i])
            m_walk = append(m_walk, m_walk[i])
            phi_walk = append(phi_walk, phi_walk[i])
            l_walk = append(l_walk, l_prime)
            #print "Dentro del else"
```

```python
max_likelihood_id = argmax(l_walk)
best_r0 = r0_walk[max_likelihood_id]
best_r1 = r1_walk[max_likelihood_id]
best_m = m_walk[max_likelihood_id]
best_phi = phi_walk[max_likelihood_id]
print log10(l_walk[max_likelihood_id])
print best_r0
print best_r1
print best_m
print best_phi
```
